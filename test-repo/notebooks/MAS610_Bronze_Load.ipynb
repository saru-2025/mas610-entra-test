```python
# Databricks notebook: MAS610_Bronze_Load
# Purpose: ingest raw CSVs for MAS610 Exercise 2
from pyspark.sql import SparkSession
from datetime import datetime

spark = SparkSession.builder.appName("MAS610_Bronze_Load").getOrCreate()

# Parameters
dbutils.widgets.text("input_dir", "/mnt/raw/mas610_ex2")
dbutils.widgets.text("silver_dir", "/mnt/silver/mas610_ex2")
input_dir  = dbutils.widgets.get("input_dir")
silver_dir = dbutils.widgets.get("silver_dir")

print(f"ðŸ“¥ Reading from {input_dir}")

# Read raw CSVs
accounts_raw   = spark.read.option("header",True).csv(f"{input_dir}/accounts.csv")
loans_raw      = spark.read.option("header",True).csv(f"{input_dir}/loans.csv")
collateral_raw = spark.read.option("header",True).csv(f"{input_dir}/collateral.csv")

print(f"âœ… Records â†’ Accounts:{accounts_raw.count()}, Loans:{loans_raw.count()}, Collateral:{collateral_raw.count()}")

# Write Bronze snapshot as Delta for reproducibility
(accounts_raw.write.mode("overwrite").format("delta").save(f"{silver_dir}/_bronze_accounts.delta"))
(loans_raw.write.mode("overwrite").format("delta").save(f"{silver_dir}/_bronze_loans.delta"))
(collateral_raw.write.mode("overwrite").format("delta").save(f"{silver_dir}/_bronze_collateral.delta"))

print("ðŸ’¾ Bronze ingestion complete.")
```
