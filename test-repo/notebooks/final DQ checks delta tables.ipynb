```python
# COMMAND ----------
# This cell executes all data checks sequentially, displaying labeled results.

from pyspark.sql import SparkSession

# --- 1. Define the Data Access Function ---
def display_delta_table_content(delta_path: str, report_title: str):
    """Reads a Delta Lake table from a specified path and displays its content."""
    
    # Use Markdown-style formatting for a clear, labeled output header
    print("\n" + "=" * 60)
    print(f"  ðŸ“Š REPORT: {report_title}")
    print("=" * 60)
    
    try:
        # Load the data using the Delta format handler
        data_frame = spark.read.format("delta").load(delta_path)
        
        # Use display() for interactive, rich-HTML output in Databricks
        display(data_frame)
        
    except Exception as e:
        # Log a warning if a table is empty or cannot be read (e.g., if the path is invalid)
        print(f"âš ï¸ FAILURE: Could not load data from path: {delta_path}")
        print(f"  Error details: {e}")
        
# --- 2. Execution Sequence (Your Pipeline Outputs) ---

# NOTE: The cluster must be running and the '/mnt/raw' mount point must be active 
# (or replaced with the full abfss:// path) for these paths to work.

# A. Final Aggregates (The main result)
display_delta_table_content(
    delta_path="/mnt/raw/risk_window_agg", 
    report_title="CORE AGGREGATES: 5-Min Window Totals"
)

# B. High-Risk Alerts
display_delta_table_content(
    delta_path="/mnt/raw/risk_alerts", 
    report_title="HIGH-RISK ALERTS: Trades > $2 Million"
)

# C. Data Quality Failures
display_delta_table_content(
    delta_path="/mnt/raw/risk_dq_results", 
    report_title="DQ RESULTS: Validation Failures"
)

# D. Dead Letter Queue
display_delta_table_content(
    delta_path="/mnt/raw/risk_dlq", 
    report_title="DLQ: Malformed Records (Raw Payloads)"
)

# COMMAND ----------
```
